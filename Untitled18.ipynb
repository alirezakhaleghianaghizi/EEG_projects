```python
import pandas as pd
file_path = '/content/svm_train.txt'
df = pd.read_csv(file_path, delimiter=' ')
df
```

```python
features = df[['x1', 'x2']].values
labels = df['y'].values
print("Features (X):")
print(features)
print("\nLabels (y):")
print(labels)
```

```python
import cvxpy as cp
import numpy as np
X =features
y = labels
n = X.shape[0]
print(n)
w = cp.Variable(2)
b = cp.Variable()
xi = cp.Variable(n)
C = 1
objective = cp.Minimize(0.5 * cp.norm(w)**2 + C * cp.sum(xi))
constraints = [y[i] * (X[i, :] @ w + b) >= 1 - xi[i] for i in range(n)]
constraints += [xi >= 0]
prob = cp.Problem(objective, constraints)
prob.solve()
optimal_w = w.value
optimal_b = b.value
print("Optimal w:")
print(optimal_w)
print("Optimal b:")
print(optimal_b)
```

```python
import cvxpy as cp
import numpy as np
C = 1.0
n = X.shape[0]
alpha = cp.Variable(n)
```

```python
Y = np.diag(y)
```

```python
X=features
P=Y@X@X.T@Y
objective = cp.Maximize(cp.sum(alpha) - 0.5 *cp.quad_form(alpha,cp.Parameter(shape=P.shape, value=P, PSD=True) ))

constraints = [0 <= alpha]
constraints+=[alpha <= C]
constraints+=[cp.sum(cp.multiply(alpha, y)) == 0]

dual_problem = cp.Problem(objective, constraints)
dual_problem.solve()

alpha_optimal = alpha.value

w_optimal = np.sum(alpha_optimal * y * X.T, axis=1)
b_optimal = np.mean(y - X @ w_optimal)

print("Optimal Lagrange multipliers (alpha):")
print(alpha_optimal)
print("\nOptimal weight vector (w):")
print(w_optimal)
print("\nOptimal bias (b):")
print(b_optimal)
```

```python
import pandas as pd
file_path = '/content/svm_train.txt'
df = pd.read_csv(file_path, delimiter=' ')
df
```

```python
features = df[['x1', 'x2']].values
labels = df['y'].values
print("Features (X):")
print(features)
print("\nLabels (y):")
print(labels)
```

```python
import cvxpy as cp
import numpy as np
X =features
y = labels
n = X.shape[0]
w = cp.Variable(2)
b = cp.Variable()
xi = cp.Variable(n)
C = 1
objective = cp.Minimize(0.5 * cp.norm(w)**2 + C * cp.sum(xi))
constraints = [y[i] * (X[i, :] @ w + b) >= 1 - xi[i] for i in range(n)]
constraints += [xi >= 0]
prob = cp.Problem(objective, constraints)
prob.solve()
optimal_w = w.value
optimal_b = b.value
print("Optimal w:")
print(optimal_w)
print("Optimal b:")
print(optimal_b)
```
